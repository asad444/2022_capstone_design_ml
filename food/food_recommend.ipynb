```python
import pandas as pd
food_data = pd.read_excel('./식품영양성분DB_음식.xlsx')

food_data = food_data.rename(columns=food_data.iloc[2])
food_data.drop(food_data.index[0:3], inplace=True)
food_data = food_data.set_index("NO")

food_data_sample = food_data[["식품명", "단백질(g)", "탄수화물(g)", "자당(g)", "철(㎍)", "마그네슘(㎎)", "아연(㎎)", 
                              "셀레늄(㎍)", "비타민 B1(㎎)", "비타민 B2(㎎)", "엽산(DFE)(㎍)", "비타민 B12(㎍)",
                             "에이코사펜타에노산(20:5(n-3))(g)", "도코사펜타에노산(22:5(n-3))(g)", "도코사헥사에노산(22:6(n-3))(g)"]]
food_data_sample["score"] = 0
food_data_sample.head()
```

```python
def plus_score(df):
    for i in range(len(df)):
        df.at[i, "score"] += i
        
    return df
```

```python
for i in food_data_sample.columns[1:-1]:
    food_data_sample = food_data_sample.sort_values(by=i, ascending=[False])
    food_data_sample = food_data_sample.reset_index()
    food_data_sample = food_data_sample.drop(food_data_sample.columns[[0]], axis=1)
    food_data_sample = plus_score(food_data_sample)

food_data_sample = food_data_sample.sort_values(by="score", ascending=[True])
food_data_sample = food_data_sample.reset_index()
food_data_sample = food_data_sample.drop(food_data_sample.columns[[0]], axis=1)

food_data_sample.head()
```

```python
food_data_sample["식품명"][0:10]
```

```python
from gensim import models
# cc.ko.300.bin -> https://fasttext.cc/docs/en/crawl-vectors.html
model = models.fasttext.load_facebook_model('F:/4-1/cc.ko.300.bin')
```

```python
keyword_list = ["산책", "호수"]

score1 = []
score2 = []

for food in food_data_sample["식품명"][0:50]:
    score1.append(model.wv.similarity(keyword_list[0], food))

for food in food_data_sample["식품명"][0:50]:
    score2.append(model.wv.similarity(keyword_list[1], food))

    
score = score1 + score2
```

```python
list_df = pd.DataFrame([score])
list_df = pd.DataFrame((zip(food_data_sample["식품명"][0:50], score)), columns = ['food', 'score'])

list_df = list_df.sort_values(by="score", ascending=[False])

list_df
```
